```python
import pandas as pd
from sklearn.datasets import fetch_california_housing
```

```python
# Loading Dataset
housing=fetch_california_housing()
```

```python
# Convert to DataFrame
df=pd.DataFrame(housing.data, columns=housing.feature_names)

# Target
df['MedHouseVal']=housing.target
```

```python
# Displaying 1st 5 rows
print(df.head())
# Displaying the shape
print('Shape of the dataset:',df.shape)
```

```python
# Checking for missing values
print('Missing values')
print(df.isnull().sum())
# Displaying Summary Statistics
print('Summary Statistics:')
print(df.describe())
# dataTypes
print('\nData Types:')
print(df.dtypes)
# Correlation matrix
print('\nCorrelation Matrix:')
print(df.corr())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop('MedHouseVal',axis=1)
Y=df['MedHouseVal']

# Splitting the Dataset
X_train,X_test, Y_train, Y_test=train_test_split(X,Y,train_size=0.8,random_state=42)

print('Training Features Shape:',X_train.shape)
print('Testing Features shape:',X_test.shape)

print('Training Target Shape:',Y_train.shape)
print('Testing Target Shape:',Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)

# fit() it trains linear Regression model by find the best coefficient
# and the intercept value that minimizes the prediction error
print('Linear Regression model trained successfully!')
```

```python
# Displaying the intercept and coefficients
print('Intercept:',lr.intercept_)

coeff_df=pd.DataFrame({'Feature':X.columns,'Coefficient':lr.coef_})
print('\n Coefficients')
print(coeff_df)
```

```python
# Now predicting the house value for all datasets
Y_pred=lr.predict(X_test)
print('Predicted values')
print(Y_pred[:10])

# Comparing the Actual vs Predicted
comparison=pd.DataFrame({'Actual':Y_test.values,'Predicted':Y_pred})
print('\nActual vs Predicted')
print(comparison.head(10))
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse=mean_squared_error(Y_test,Y_pred)
r2=r2_score(Y_test,Y_pred)

print('Mean Squared Error(MSE):',mse)
print('R2 Score:',r2)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
splits=[0.6,0.7,0.8,0.9]
results=[]
for trainsize in splits:
  # Split Data
  X_train, X_test, Y_train, Y_test=train_test_split(X,Y,train_size=trainsize,random_state=42)

  # Train model
  lr=LinearRegression()
  lr.fit(X_train, Y_train)

  # Predict
  Y_pred=lr.predict(X_test)

  # Evaluate
  mse=mean_squared_error(Y_test, Y_pred)
  r2=r2_score(Y_test,Y_pred)

  # Storing the result
  results.append([f"{round(trainsize*100)}:{round((1-trainsize)*100)}",mse,r2])

# Displaying the Results
results_df=pd.DataFrame(results,columns=['Train:Test Split','MSE','R2 Score'])
print(results_df)
```

```python
from sklearn.linear_model import Ridge
```

```python
# Part:A Ridge Regression
alphas=[0.001, 0.005,0.01,0.05,0.1,0.5]
splits=[0.6,0.7,0.8,0.9]
ridge_results=[]
for trainsize in splits:
  # Split Data
  X_train, X_test, Y_train, Y_test=train_test_split(X,Y,train_size=trainsize,random_state=42)

  for alp in alphas:
    ridge=Ridge(alpha=alp)
    ridge.fit(X_train,Y_train)
    Y_pred=ridge.predict(X_test)
    mse=mean_squared_error(Y_test, Y_pred)
    r2=r2_score(Y_test, Y_pred)
# Storing the Result
    ridge_results.append([f"{round(trainsize*100)}:{round((1-trainsize)*100)}",mse,r2])

ridge_df=pd.DataFrame(ridge_results,columns=['Train:Test Split','MSE','R2 Score'])
print(ridge_df)
```

```python
# Part:B Lasso Regression
from sklearn.linear_model import Lasso
```

```python
lasso_results=[]
for trainsize in splits:
  # Split Data
  X_train, X_test, Y_train, Y_test=train_test_split(X,Y,train_size=trainsize,random_state=42)
  for alp in alphas:
    lasso=Lasso(alpha=alp)
    lasso.fit(X_train, Y_train)
    Y_pred=lasso.predict(X_test)

    mse=mean_squared_error(Y_test,Y_pred)
    r2=r2_score(Y_test, Y_pred)

    lasso_results.append([f'{round(trainsize*100)}:{round((1-trainsize)*100)}',alp,mse,r2])

lasso_df=pd.DataFrame(lasso_results,columns=['Train: Train Split','Alpha','Mse','R2 Score'])

print(lasso_df)
```
