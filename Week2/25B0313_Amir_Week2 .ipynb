```python
import numpy as np
marks = np.array([
    [78, 82, 85],
    [91, 89, 94],
    [67, 72, 70],
    [88, 90, 86],
    [56, 60, 58]
])
student_names = ["Aarav", "Diya", "Rohan", "Ananya", "Kabir"]
```

```python
print("Shape of the array:", marks.shape)
print("Number of dimensions:", marks.ndim)
print("Data type:", marks.dtype)
```

```python
# import numpy as np
student_names = ["Aarav", "Diya", "Rohan", "Ananya", "Kabir"]
print("Total marks of each student:")
totalmarks=np.sum(marks,axis=1)
for i in range(len(student_names)):
  print(student_names[i],":",totalmarks[i])
print("Average marks for each student:")
avg=np.mean(marks,axis=1)
for i in range(len(student_names)):
  print(f"{student_names[i]} : {avg[i]:.2f}")
```

```python
subject_names = ["Math", "Science", "English"]
print("Highest marks in each subject:")
highest_marks=np.max(marks,axis=0)
for i in range(len(subject_names)):
  print(student_names[i],":",highest_marks[i])

print("Lowest marks in each subject:")
lowest_marks=np.min(marks,axis=0)
for i in range(len(subject_names)):
  print(student_names[i],":",lowest_marks[i])
```

```python
# Adding 5 grace marks to each student
grace_marks=marks+5
print("Marks after adding 5 grace marks:")
print(grace_marks)
```

```python
# Now replacing all the marks less than 70 with 0
newmarks=np.where(marks<70,0,marks)
print("Marks after repalacing value below 70 with 0",newmarks)
```

```python
# To find the topper student with highest total marks
topper=np.argmax(totalmarks)
# print("Topper student:",student_names[topper])
# or
print("Student ",topper+1,"is the topper","and the name of the student is ",student_names[topper])
```

```python
# Create a new array containing students whose average marks are above 80
highavg=np.array(student_names)[avg>80]
print("Students with average above 80:")
print(highavg)
```

```python
# Count how many marks are greater than 85
high85=np.sum(marks>85)
print("Number of marks greater than 85:",high85)
```


# 3. Matplotlib Basics

Matplotlib is used for plotting and visualizing data.


```python
# Note: Make sure you have completed the Pandas section above
# so that the Total_Marks column exists in the dataframe.
# Line plot of student averages

plt.figure(figsize=(8, 5))

plt.plot(df["Name"], df["Average"], marker='o')

plt.title("Student Average Scores")
plt.xlabel("Student Name")
plt.ylabel("Average Score")

plt.xticks(rotation=45)
plt.grid(True)

plt.show()
```

```python

# Bar chart

subject_means = [
    df["Math"].mean(),
    df["Science"].mean(),
    df["English"].mean()
]

subjects = ["Math", "Science", "English"]

plt.bar(subjects, subject_means)

plt.title("Average Marks by Subject")
plt.ylabel("Average Marks")

plt.show()
```

```python

# Histogram

all_marks = pd.concat([
    df["Math"],
    df["Science"],
    df["English"]
])

plt.hist(all_marks, bins=10)

plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('final_assignment_dataset.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
df['Total_Marks']=df['Math']+df['Science']+df['English']
df['Average_Marks']=df['Total_Marks']/3
# print(df)
highest_student=df.loc[df['Total_Marks'].idxmax()]
lowest_student=df.loc[df['Total_Marks'].idxmin()]


print("Highest scoring student:")
print(highest_student)
print("Lowest scoring student:")
print(lowest_student)

print("Students scoring above 80 average:")
print(df[df['Average_Marks']>80])
```

```python
maths_marks=np.array(df['Math'])
science_marks=np.array(df['Science'])
english_marks=np.array(df['English'])

maths_avg=np.mean(maths_marks)
science_avg=np.mean(science_marks)
english_avg=np.mean(english_marks)

print("Average marks in Maths:",maths_avg)
print("Average marks in Science:",science_avg)
print("Average marks in English:",english_marks)

print("Median of Maths",np.median(maths_marks))
print("Median of Science",np.median(science_marks))
print("Median of English",np.median(english_marks))

print("Standard Deviation of Maths",np.std(maths_marks))
print("Standard Deviation of Science",np.std(science_marks))
print("Standard Deviation of English",np.std(english_marks))

average_marks=[maths_avg,science_avg,english_avg]
subjects=['Maths','Science','English']
print("Subject with the highest average marks:")
highes_average=np.argmax(average_marks)
print(subjects[highes_average])
```

```python
from matplotlib.lines import lineStyles
# Line Plot of Student Average Marks
plt.figure(figsize=(13,4))
plt.plot(df['Name'],df['Average_Marks'],marker='o',linestyle=':',color='blue')
plt.title('Student Average Marks')
plt.xlabel('Student Name')
plt.ylabel('Average Marks')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Histogram of Marks
plt.figure(figsize=(8,5))
all_marks = pd.concat([df["Math"], df["Science"], df["English"]])

plt.hist(all_marks,bins=9,color='blue',edgecolor='red')
plt.title("Distribution of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Bar Chart of Subject-wise Averages
maths_avg=np.mean(maths_marks)
science_avg=np.mean(science_marks)
english_avg=np.mean(english_marks)

subject=['Math','Science','English']
averages=[maths_avg,science_avg,english_avg]
plt.bar(subject,averages)

plt.title('Subject-wise Averages')
plt.xlabel('Subject')
plt.ylabel('Average Marks')
plt.show()
```
