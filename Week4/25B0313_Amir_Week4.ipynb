#Exercise 4:
Explain in your own words the difference between One-vs-Rest and Softmax
(multinomial) approaches to multiclass classification.

Answer:


* One-vs-Rest:This strategy trains a separate binary classifier for
each individual class. In this each classifier is trained in complete isolation, hence it has no idea that it will be compared against other classifiers. This isolation hurts overall accuracy and creates a blind spot where the combined model fails to predict a class even though a perfect separator exists.



* Softmax Regression: Instead of building separate classifiers, this is a direct approach trains the parameters for all the classed jointly in a single model.On adding the probability distribution of all the classes it non-negative and is equal to 1, because it learns all together and naturally avoids blind spot problem which in the case for One-vs-Rest.

```python
# Exercise 5
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression

# Loading data and training the model
data = load_breast_cancer()
X, y = data.data, data.target
model = LogisticRegression(max_iter=10000)
model.fit(X, y)

# First 5 test samples
X_test = X[:5]

# Print predicted probabilities
# This outputs the probability that each sample belongs to Class 1
probabilities = model.predict_proba(X_test)[:, 1]
print("Predicted Probabilities (P=1):", probabilities)

# 3. Print hard predictions with the default threshold (alpha = 0.5)
predictions_default = model.predict(X_test)
print("Predictions (alpha=0.5):", predictions_default)

# 4. Try custom thresholds (alpha = 0.3 and alpha = 0.7)
predictions_03 = (probabilities >= 0.3).astype(int)
predictions_07 = (probabilities >= 0.7).astype(int)

print("Predictions (alpha=0.3):", predictions_03)
print("Predictions (alpha=0.7):", predictions_07)
```
